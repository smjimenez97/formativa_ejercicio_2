# Clustering con el Método K-Means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

#%matplotlib inline
```

Importamos los datos

```python
vinos=pd.read_csv("/content/caracteristicas de vinos.csv", engine="python")
```

Visualizamos los datos

```python
vinos.info()
```

Analizamos las primeras filas

```python
vinos.head()
```

Se elimina la columna Vino, ya que es la que se va a analizar.

```python
vinos_variables=vinos.drop(["Vino"], axis=1)
```

Analizamos los valores minimos, máximos, promedio, desviación estandar y cuartiles de cada columna.

```python
vinos_variables.describe()
```

Normalizar valores

```python
vinos_norm = (vinos_variables - vinos_variables.min()) / (vinos_variables.max() - vinos_variables.min())
vinos_norm
```

```python
vinos_norm.describe()
```

Busqueda de la cantidad óptima de clusters

Analizamos la similutd de los diferentes datos dentro de los clusters

```python
wcss = []

for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, max_iter = 300)
  kmeans.fit(vinos_norm) #Se aplica K-Means a los datos
  wcss.append(kmeans.inertia_)
```

Grafico de los resultados de WCSS para formar el codo de Jambú

```python
plt.plot(range(1,11), wcss)
plt.title("Codo de Jambú")
plt.xlabel("Número de Clusters")
plt.ylabel("WCSS") #WCSS. es un indicador que muestra la similitud de los datos dentro de los clusters
plt.show()
```

Aplicamos el método k-means a la base de datos

```python
clustering = KMeans(n_clusters = 3, max_iter = 300) #Crea el modelo
clustering.fit(vinos_norm) # Aplicamos el modelo a la base de datos
```

Agregando la clasificación al archivo original

```python
vinos["KMeans_Clusters"] = clustering.labels_
vinos.head()
```

Aplicamos PCA (Análisis de Componentes Principales) para observar como se conforman los clusters

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_vinos = pca.fit_transform(vinos_norm)
pca_vinos_df = pd.DataFrame(data = pca_vinos, columns = ["Componente_1", "Componente_2"])
pca_nombres_vinos = pd.concat([pca_vinos_df, vinos[["KMeans_Clusters"]]], axis=1)

pca_nombres_vinos
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel("Componente 1", fontsize = 15)
ax.set_ylabel("Componente 2", fontsize = 15)
ax.set_title("Componentes Principales", fontsize = 20)

ax.scatter(x = pca_nombres_vinos.Componente_1, 
           y = pca_nombres_vinos.Componente_2, 
           c = pca_nombres_vinos.KMeans_Clusters,
           s=50)
plt.show()
```
